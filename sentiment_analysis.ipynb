## Step 1. VADER Sentiment Scoring

use NLTK's `SentimentIntensityAnalyzer` to get neg/neu/pos scores of the text
- bag of words approach:
  1. stop words are removed (the/and/etc.)
  2. each word is scored and combined to a total score.

VADER stands for valence aware dicionary and sEntiment Reasoner
-doesn't account for relations between words

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('I am so happy!')
```

```python
sia.polarity_scores('This is the worst thing ever.')
```

```python
sia.polarity_scores(example)
```

```python
# Run the polarity score on the entire dataset
res = {}

for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
    
```

```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={'index' : 'Id'})
vaders = vaders.merge(df, how='left')
```

```python
# Now we have sentiment score and metadata
vaders.head()
```

## Plot VADER results

```python
ax = sns.barplot(data=vaders, x='Score', y='compound')
ax.set_title('Compound Score by Amazon Star Review')
plt.show() # expected !
```

## Step 3. Roberta Pretrained Model
- Use a model trained of a large corpus of data.
- Transformer model accounts for the words but also the context related to other words.

```python
from transformers import AutoTokenizer # similar to nltk
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment" 
# provided by huggingface - pretrained model for sentiment analysis (have own model weights)
# trained on labeled twitter comments - downloading weights:
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
# remember prior example from set (negative about oatmeal)
print(example)
sia.polarity_scores(example)
```

```python
# Run for Roberta Model - put example into 1 & 0s
encoded_text = tokenizer(example, return_tensors='pt')
output = model(**encoded_text)
scores = output[0][0].detach().numpy()
scores = softmax(scores)
scores_dict = {
    'roberta_neg' : scores[0],
    'roberta_neu' : scores[1],
    'roberta_pos' : scores[2]
}
print(scores_dict) # more powerful bc it considers more negative than vaders model
```

```python
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['Text']
        myid = row['Id']
        vader_result = sia.polarity_scores(text)
        vader_result_rename = {}
        for key, value in vader_result.items(): # rename vader
            vader_result_rename[f"vader_{key}"] = value
        roberta_result = polarity_scores_roberta(text)
        both = {**vader_result_rename, **roberta_result}
        res[myid] = both
    except RuntimeError: # stating that the roberta model breaks for examples where text is too long -> print in some way
        print(f'Broke for id {myid}')
```

```python
results_df = pd.DataFrame(res).T
results_df = results_df.reset_index().rename(columns={'index' : 'Id'})
results_df = results_df.merge(df, how='left')
```
